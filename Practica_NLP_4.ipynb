####Cargar el Modelo LSTM


```python
from google.colab import files

# Subir los archivos necesarios desde tu máquina local
uploaded = files.upload()

# Verifica los archivos subidos
import os
print("Archivos subidos:", os.listdir())

# Cargar el modelo de LSTM
from keras.models import load_model
model = load_model('sentiment_lstm_model_deeplearning.keras')

# Cargar los resultados del entrenamiento desde el archivo CSV
import pandas as pd
training_results = pd.read_csv('training_results_lstm_deeplearning.csv')

# Verificar el contenido de los resultados de entrenamiento
print(training_results.head())
```

##Gráficas de Exactitud (Accuracy) y Pérdida (Loss)

```python
import matplotlib.pyplot as plt

# Gráfica de exactitud (accuracy)
plt.figure(figsize=(12, 6))
plt.plot(training_results['epoch'], training_results['train_accuracy'], label='Train Accuracy')
plt.plot(training_results['epoch'], training_results['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

# Gráfica de pérdida (loss)
plt.figure(figsize=(12, 6))
plt.plot(training_results['epoch'], training_results['train_loss'], label='Train Loss')
plt.plot(training_results['epoch'], training_results['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

```

#Conclusiones sobre el modelo LSTM

- Rendimiento actual: El modelo alcanzó una exactitud de 88.36% en entrenamiento y 80.11% en prueba, con pérdidas de 0.3213 y 0.4970, respectivamente. Aunque el modelo muestra buen desempeño, existe una ligera diferencia entre las métricas de entrenamiento y prueba, indicando posibles signos de sobreajuste.
- Aspectos positivos: Captura patrones secuenciales con eficacia y generaliza adecuadamente en datos de prueba.
- Limitaciones: La diferencia entre la pérdida y exactitud en los conjuntos de entrenamiento y prueba podría ser indicativa de un modelo que podría beneficiarse de mayor regularización o más datos.

##Recomendaciones

###Optimización del modelo:
- Probar con diferentes tasas de regularización (Dropout, L2) para reducir el sobreajuste.
- Ajustar hiperparámetros como la tasa de aprendizaje y el número de unidades LSTM.

###Ampliar y diversificar datos:
- Incrementar el tamaño del conjunto de datos mediante recopilación o aumentación.
- Verificar balance de clases para evitar sesgos.

###Enriquecer el preprocesamiento:
- Revisar la longitud máxima de las secuencias y el vocabulario (max_words) para representar mejor el texto.
- Utilizar embeddings preentrenados como GloVe o Word2Vec para mejorar la semántica.

###Explorar arquitecturas alternativas:
- Combinar LSTM con capas GRU.
- Probar modelos más avanzados como Transformers (ej., BERT o GPT).

##Conclusión Final

El modelo LSTM presenta un rendimiento sólido y competitivo con una exactitud superior al 80% en datos de prueba. Sin embargo, puede mejorarse ajustando hiperparámetros, ampliando los datos y explorando arquitecturas híbridas o más avanzadas.
